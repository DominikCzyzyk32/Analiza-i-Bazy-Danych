```python
# Zaimportowanie odpowiednich bibliotek
import pandas as pd
```

```python
# Wczytanie danych z pliku csv
df = pd.read_csv('../OriginalData/tb.csv')
```

```python
# Początkowe, poglądowe wyświetlenie danych
df
```

```python
# Zmiana nazw większości kolumn (zmiennych), tak aby lepiej opisywały rodzaj przechowywanych danych
df_renamed = df.rename(columns = {'iso2':'country', 'new_sp':'total_cases', 'new_sp_m04': 'm0-4', 'new_sp_m514': 'm5-14', 
                                  'new_sp_m014': 'm0-14', 'new_sp_m1524': 'm15-24', 'new_sp_m2534': 'm25-34', 
                                  'new_sp_m3544': 'm35-44', 'new_sp_m4554': 'm45-54', 'new_sp_m5564': 'm55-64', 
                                  'new_sp_m65': 'm65', 'new_sp_mu': 'm65+','new_sp_f04': 'f0-4', 'new_sp_f514': 'f5-14', 
                                  'new_sp_f014': 'f0-14', 'new_sp_f1524': 'f15-24', 'new_sp_f2534': 'f25-34', 
                                  'new_sp_f3544': 'f35-44', 'new_sp_f4554': 'f45-54', 'new_sp_f5564': 'f55-64', 
                                  'new_sp_f65': 'f65', 'new_sp_fu': 'f65+' })
df_renamed 
```

```python
# Uporządkowanie danych przy pomocy funkcji melt() (skrócenie tabeli na szerokość, czego następstwem jest jej wydłużenie, 
# poprawia to czytelność danych)
df_renamed_melted = pd.melt(df_renamed, id_vars=['country','year'], value_vars=['m0-4', 'm5-14', 'm0-14', 'm15-24','m25-34','m35-44','m45-54','m55-64','m65','m65+','f0-4', 'f5-14', 'f0-14', 'f15-24','f25-34','f35-44','f45-54','f55-64','f65','f65+'],var_name='sex/age_group', value_name='case_count')
df_renamed_melted
```

```python
# Rozbicie kolumny 'sex/age_group' na dwie osobne kolumny 'sex' i 'age group'
sex_tab = []
age_group_tab = []

for i in range(df_renamed_melted.shape[0]):
    sex_tab.append(df_renamed_melted['sex/age_group'][i][0])
    age_group_tab.append(df_renamed_melted['sex/age_group'][i][1:])

df_renamed_melted.drop(['sex/age_group'], axis=1, inplace=True)
df_renamed_melted.insert(2, 'sex', sex_tab)
df_renamed_melted.insert(3, 'age_group', age_group_tab)

df_renamed_melted
```

```python
# Usunięcie rekordów zawierających niezidentyfikowane dane (Nan)
df_renamed_melted_without_nans = df_renamed_melted.dropna(subset = ['country', 'case_count'])
df_renamed_melted_without_nans.reset_index(drop=True)
```

```python
# Posortowanie danych po kraju i roku przeprowadzenia badania
df_renamed_melted_without_nans_sorted = df_renamed_melted_without_nans.sort_values(['country', 'year'])
df_renamed_melted_without_nans_sorted.reset_index(drop=True)
```

```python
# Zapisanie uporządkowanych danych do pliku csv w katalogu /AnalysisData
df_renamed_melted_without_nans_sorted.to_csv('../AnalysisData/processed_tb.csv', index=False)
```
