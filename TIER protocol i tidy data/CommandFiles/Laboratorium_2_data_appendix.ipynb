```python
# Zaimportowanie odpowiednich bibliotek
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Wczytanie danych z pliku csv (tych już uporządkowanych)
df = pd.read_csv('../AnalysisData/processed_tb.csv')
```

```python
# Podstawowe informacje o danych
df.info(verbose=True)
```

```python
# Stworzenie wykresu słupkowego opisującego liczbę zachorowań na gruźlicę w poszczególnych latach
years = df['year'].unique()
num_of_cases_in_given_years = []
for year in years:
    num_of_cases_in_given_years.append(df[df['year'].isin([year])].sum()['case_count'])

plt.figure(figsize=(10,10))
plt.bar(years, num_of_cases_in_given_years)
plt.title("Liczba zachorowań na gruźlicę w poszczególnych latach")
plt.xlabel("Rok")
plt.ylabel("Liczba zachorowań")
plt.show()
```

```python
# Stworzenie wykresu słupkowego opisującego liczbę zachorowań na gruźlicę wg. płci w latach 1980-2008
sexs = df['sex'].unique()
num_of_cases_by_sexs = []
for sex in sexs:
    num_of_cases_by_sexs.append(df[df['sex'].isin([sex])].sum()['case_count'])

plt.figure(figsize=(10,10))
plt.bar(sexs, num_of_cases_by_sexs)
plt.title("Liczba zachorowań na gruźlicę wg. płci w latach 1980-2008")
plt.xlabel("Płeć")
plt.ylabel("Liczba zachorowań")
plt.show()
```

```python
# Stworzenie wykresu słupkowego opisującego liczbę zachorowań na gruźlicę poszczególnych grup wiekowych w latach 1980-2008
age_groups = ['0-4','5-14','0-14','15-24','25-34','35-44','45-54','55-64','65','65+']
num_of_cases_by_age_groups = []
for age_group in age_groups:
    num_of_cases_by_age_groups.append(df[df['age_group'].isin([age_group])].sum()['case_count'])
    
plt.figure(figsize=(10,10))
plt.bar(age_groups, num_of_cases_by_age_groups)
plt.title("Liczba zachorowań na gruźlicę poszczególnych grup wiekowych w latach 1980-2008")
plt.xlabel("Grupa wiekowa")
plt.ylabel("Liczba zachorowań")
plt.show()
```
