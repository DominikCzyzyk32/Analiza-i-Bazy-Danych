```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Zadanie 3

```python
# Zdefiniowanie funkcji
def f(x):
    return x**2 + 5
```

```python
# Przygotowanie trzech zestaw√≥w argument√≥w
x_1 = np.linspace(-1, 1)
x_2 = np.linspace(-6, 6)
x_3 = np.linspace(0, 5)

# Wyrysowanie wykres√≥w zdefiniowanej wy≈ºej funkcji dla r√≥≈ºnych zestaw√≥w argument√≥w
plt.figure()
plt.plot(x_1, f(x_1), label='x^2 + 5')
plt.title("Wykres funkcji x^2 + 5 na przedziale [-1, 1]")
plt.xlabel("o≈õ x")
plt.ylabel("o≈õ y")
plt.legend()
plt.show()
```

```python
plt.figure()
plt.plot(x_2, f(x_2), label='x^2 + 5')
plt.title("Wykres funkcji x^2 + 5 na przedziale [-6, 6]")
plt.xlabel("o≈õ x")
plt.ylabel("o≈õ y")
plt.legend()
plt.show()
```

```python
plt.figure()
plt.plot(x_3, f(x_3), label='x^2 + 5')
plt.title("Wykres funkcji x^2 + 5 na przedziale [0, 5]")
plt.xlabel("o≈õ x")
plt.ylabel("o≈õ y")
plt.legend()
plt.show()
```
